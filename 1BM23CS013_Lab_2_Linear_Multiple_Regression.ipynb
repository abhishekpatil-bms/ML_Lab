```python
# USING SKLEARN

import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

print(df.head())

# Define X and y
X = df[['year']]
y = df['per capita income (US$)']

# Create model
model = LinearRegression()
model.fit(X, y)

# Predict for 2020
prediction = model.predict([[2020]])

print("\n--- Using Library ---")
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
print("Predicted Per Capita Income in 2020:", prediction[0])
```

```python
# WITHOUT SKLEARN

import pandas as pd
import numpy as np

df = pd.read_csv("canada_per_capita_income.csv")

X = df['year'].values
y = df['per capita income (US$)'].values

# Add column of 1s for intercept
X = np.column_stack((np.ones(len(X)), X))

# Apply Normal Equation
beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print("\n--- Without Library ---")
print("Intercept:", beta[0])
print("Slope:", beta[1])

# Predict for 2020
X_new = np.array([1, 2020])
prediction = X_new.dot(beta)

print("Predicted Per Capita Income in 2020:", prediction)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Load dataset
df = pd.read_csv("salary.csv")

# Define X and y
X = df[['YearsExperience']]
y = df['Salary']

# Create imputer (replace NaN with mean)
imputer = SimpleImputer(strategy='mean')

X = imputer.fit_transform(X)

# Build model
model = LinearRegression()
model.fit(X, y)

# Predict for 12 years
prediction = model.predict([[12]])

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
print("Predicted Salary for 12 years:", prediction[0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("salary.csv")

X = df['YearsExperience'].values
y = df['Salary'].values

# Replace NaN with mean manually
mean_value = np.nanmean(X)
X = np.where(np.isnan(X), mean_value, X)

# Add column of 1s
X = np.column_stack((np.ones(len(X)), X))

# Normal Equation
beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print("Intercept:", beta[0])
print("Slope:", beta[1])

# Predict for 12 years
X_new = np.array([1, 12])
prediction = X_new.dot(beta)

print("Predicted Salary for 12 years:", prediction)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Load dataset
df = pd.read_csv("hiring.csv")

print(df.head())

# Convert text numbers to numeric
mapping = {
    'zero':0, 'one':1, 'two':2, 'three':3,
    'four':4, 'five':5, 'six':6,
    'seven':7, 'eight':8, 'nine':9, 'ten':10
}

df['experience'] = df['experience'].map(mapping)

# Convert to numeric (in case of mixed types)
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

# Define X and y
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Build model
model = LinearRegression()
model.fit(X, y)

print("\nIntercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Predictions
print("\nSalary for (2,9,6):", model.predict([[2,9,6]])[0])
print("Salary for (12,10,10):", model.predict([[12,10,10]])[0])
```

```python
import numpy as np

df = pd.read_csv("hiring.csv")

# Convert text to numeric
mapping = {
    'zero':0, 'one':1, 'two':2, 'three':3,
    'four':4, 'five':5, 'six':6,
    'seven':7, 'eight':8, 'nine':9, 'ten':10
}

df['experience'] = df['experience'].map(mapping)
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']].values
y = df['salary($)'].values

# Replace NaN with column mean
col_means = np.nanmean(X, axis=0)
inds = np.where(np.isnan(X))
X[inds] = np.take(col_means, inds[1])

# Add intercept column
X = np.column_stack((np.ones(len(X)), X))

# Normal Equation
beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print("\nIntercept:", beta[0])
print("Coefficients:", beta[1:])

# Predictions
new1 = np.array([1,2,9,6])
new2 = np.array([1,12,10,10])

print("\nSalary for (2,9,6):", new1.dot(beta))
print("Salary for (12,10,10):", new2.dot(beta))
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("hiring.csv")

# Convert words to numbers
mapping = {
    'zero':0, 'one':1, 'two':2, 'three':3,
    'four':4, 'five':5, 'six':6,
    'seven':7, 'eight':8, 'nine':9, 'ten':10
}

df['experience'] = df['experience'].map(mapping)
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

# Features and target
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']].values
y = df['salary($)'].values

# Replace NaN with column mean
col_means = np.nanmean(X, axis=0)
inds = np.where(np.isnan(X))
X[inds] = np.take(col_means, inds[1])

# Convert to float (IMPORTANT FIX)
X = X.astype(float)
y = y.astype(float)

# Add intercept
X = np.column_stack((np.ones(len(X)), X))

# Normal Equation
beta = np.linalg.inv(X.T @ X) @ X.T @ y

print("Intercept:", beta[0])
print("Coefficients:", beta[1:])

# Predictions
new1 = np.array([1,2,9,6])
new2 = np.array([1,12,10,10])

print("Salary for (2,9,6):", new1 @ beta)
print("Salary for (12,10,10):", new2 @ beta)
```
